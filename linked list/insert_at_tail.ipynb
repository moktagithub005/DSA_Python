```python
# common.py (Your existing code)
class Node:
    def __init__(self, value):
        self.data = value
        self.next = None

def print_LL(head):
    temp = head
    while(temp != None):
        print(temp.data, end="->")
        temp = temp.next
    print()
    return

# return a head to a newly created LL
def take_input():
    value = int(input("Enter the value of Node :- "))
    head = None
    
    while(value != -1):
        newNode = Node(value)
        if(head == None):
            head = newNode
        else:
            temp = head
            while(temp.next != None):
                temp = temp.next
            
            temp.next = newNode
        
        value = int(input("Enter the value of Node :- "))
    
    return head

def take_input_better():
    value = int(input("Enter the value of Node :- "))
    head = None
    tail = None
    
    while(value != -1):
        newNode = Node(value)
        if(head == None):
            head = newNode
            tail = newNode
        else:
            tail.next = newNode
            tail = newNode
        
        value = int(input("Enter the value of Node :- "))
    
    return head

# --- New Function: insert_at_tail ---
def insert_at_tail(head, data):
    """
    Inserts a new node with the given data at the tail of the linked list.

    Args:
        head: The head node of the linked list.
        data: The data to be stored in the new node.

    Returns:
        The head of the modified linked list.
    """
    new_node = Node(data)

    # Edge case: If the linked list is empty, the new node becomes the head.
    if head is None:
        return new_node

    # Traverse to the last node (tail) of the linked list.
    # We use a temporary pointer `temp` to avoid modifying the original `head`.
    temp = head
    while temp.next is not None:
        temp = temp.next
    
    # At this point, `temp` is the last node.
    # Link the new node to the `next` of the current tail.
    temp.next = new_node

    # Return the original head of the list.
    return head

# Example Usage (You can uncomment this in your main script to test)
if __name__ == '__main__':
    print("--- Testing insert_at_tail ---")

    # Test 1: Inserting into an empty list
    print("\nTest Case 1: Inserting into an empty list")
    my_head = None
    my_head = insert_at_tail(my_head, 100)
    print("List after inserting 100:", end=" ")
    print_LL(my_head) # Expected: 100->

    # Test 2: Inserting into an existing list
    print("\nTest Case 2: Inserting into an existing list")
    print("Please enter values for the initial list (e.g., 10 20 30 -1):")
    initial_list_head = take_input_better() # Using take_input_better for efficiency
    print("Original List:", end=" ")
    print_LL(initial_list_head)

    print("\nInserting 40 at tail...")
    initial_list_head = insert_at_tail(initial_list_head, 40)
    print("List after inserting 40:", end=" ")
    print_LL(initial_list_head) # Expected to have 40 at the end

    print("\nInserting 50 at tail...")
    initial_list_head = insert_at_tail(initial_list_head, 50)
    print("List after inserting 50:", end=" ")
    print_LL(initial_list_head) # Expected to have 50 at the very end
```
