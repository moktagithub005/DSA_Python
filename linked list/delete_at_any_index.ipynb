```python
class Node:
    def __init__(self, data):
        self.data = data
        self.next = None


# Utility function to print the linked list
def print_list(head):
    temp = head
    while temp is not None:
        print(temp.data, end=" -> ")
        temp = temp.next
    print("None")


# Utility function to insert node at the tail (for testing)
def insert_at_tail(head, data):
    new_node = Node(data)
    if head is None:
        return new_node
    temp = head
    while temp.next is not None:
        temp = temp.next
    temp.next = new_node
    return head


# Function to delete node at a specific index
def delete_at_index(head, index):
    if head is None:
        print("LL is Empty")
        return head

    # Case 1: delete at head
    if index == 0:
        return head.next

    temp = head
    count = 0

    # Move temp to (index-1)-th node
    while temp is not None and count < index - 1:
        temp = temp.next
        count += 1

    # If index is out of bounds
    if temp is None or temp.next is None:
        print("Out of Bounds")
        return head

    # Skip the node at `index`
    nodeToBeDeleted = temp.next
    nodeAfterDeletedNode = nodeToBeDeleted.next
    temp.next = nodeAfterDeletedNode

    return head


# -----------------------------
# Example usage
# -----------------------------

# Create Linked List: 1 -> 2 -> 3 -> 4 -> 5
head = None
for i in range(1, 6):
    head = insert_at_tail(head, i)

print("Original Linked List:")
print_list(head)

# Delete at index 2 (0-based) â†’ deletes "3"
head = delete_at_index(head, 2)

print("\nAfter deleting index 2:")
print_list(head)

# Delete at head (index 0) â†’ deletes "1"
head = delete_at_index(head, 0)

print("\nAfter deleting head (index 0):")
print_list(head)

# Try deleting out of bounds (index 10)
head = delete_at_index(head, 10)

print("\nAfter trying to delete index 10 (out of bounds):")
print_list(head)
```

## âš™ï¸ Function Breakdown

### `Node` Class
The `Node` class is the fundamental building block of the linked list. Each `Node` object contains two attributes: `data`, which stores the value, and `next`, a pointer to the next node in the sequence. If `next` is `None`, it signifies the end of the list.

### `delete_at_index(head, index)`
This function is responsible for removing a node from the linked list at a given zero-based `index`. 

1.  **Empty List Check:** The function first checks if the `head` is `None`. If the list is empty, it prints an error message and returns the `head` as is.
2.  **Delete at Head:** If the `index` is `0`, the function simply returns `head.next`, effectively making the second node the new head and discarding the original first node.
3.  **Traverse to the Predecessor:** For any other index, the function initializes a `temp` pointer to the `head` and a `count` to `0`. It then iterates through the list until `temp` points to the node **just before** the one to be deleted (i.e., at `index - 1`).
4.  **Out-of-Bounds Check:** After the loop, it checks if `temp` or `temp.next` is `None`. This condition is true if the requested `index` is greater than or equal to the length of the list, indicating an out-of-bounds error. It prints a message and returns without making any changes.
5.  **Node Deletion:** If the index is valid, it reassigns `temp.next` to `temp.next.next`. This operation effectively "skips over" the node at the specified index, removing it from the chain. The garbage collector will eventually reclaim the memory of the skipped node.

### ðŸ’¡ Step-by-Step Explanation

Let's use a linked list `A -> B -> C -> D -> None` and assume we want to delete node `C`. The `temp` variable in the program would have already traversed to node `B`, which is the node right before `C`.

1.  **`nodeToBeDeleted = temp.next`**
    * `temp.next` points to the node `C`.
    * This line creates a new variable, `nodeToBeDeleted`, and assigns it the reference to node `C`. This is done for clarity, holding a temporary reference to the node we're about to remove.
    * The list remains unchanged: `A -> B -> C -> D -> None`.
2.  **`nodeAfterDeletedNode = nodeToBeDeleted.next`**
    * `nodeToBeDeleted` holds the reference to node `C`. `nodeToBeDeleted.next` points to node `D`.
    * This line creates another temporary variable, `nodeAfterDeletedNode`, and assigns it the reference to node `D`. This variable holds a reference to the node that will come **after** the one we are deleting.
    * The list is still unchanged: `A -> B -> C -> D -> None`.
3.  **`temp.next = nodeAfterDeletedNode`**
    * `temp` is still at node `B`. `temp.next` is the pointer from `B` to `C`.
    * `nodeAfterDeletedNode` holds the reference to node `D`.
    * This line is the key to the deletion. It reassigns the pointer from `B` to `D`.
    * After this line, the linked list looks like this: `A -> B -> D -> None`. The node `C` is no longer part of the list's chain.

By changing `temp.next` to point directly to `nodeAfterDeletedNode`, the code effectively **skips** over `nodeToBeDeleted`. The `nodeToBeDeleted` (node `C`) is now orphaned and, in most languages with automatic memory management like Python, will eventually be cleaned up by the garbage collector.
