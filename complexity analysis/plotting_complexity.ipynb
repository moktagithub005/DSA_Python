```python
import time
import numpy as np
import matplotlib.pyplot as plt

# --- 1. Bubble Sort Implementation for Time Measurement ---

def bubble_sort(arr):
    """
    Implements the Bubble Sort algorithm.
    This version is for demonstration of time measurement and comparison
    with theoretical O(n^2) behavior in the worst case.
    """
    n = len(arr)
    # Outer loop for passes (n-1 passes are sufficient)
    for i in range(n):
        # Inner loop for comparisons and swaps.
        # After each pass 'i', the largest 'i' elements are at their correct end positions.
        for j in range(0, n - i - 1):
            # Compare adjacent elements and swap if they are in the wrong order
            if arr[j] > arr[j + 1]:
                arr[j], arr[j + 1] = arr[j + 1], arr[j]

def measure_time_bubble_sort(input_sizes):
    """
    Measures the time taken for the Bubble Sort algorithm for different input sizes.
    It generates reverse-sorted arrays to simulate the worst-case scenario for Bubble Sort.
    """
    times = []
    print("\n--- Measuring Time for Bubble Sort ---")
    for size in input_sizes:
        # Create a reverse-sorted array (worst case for Bubble Sort)
        # Convert numpy array to list as bubble_sort expects a list for in-place modification
        arr = np.arange(size, 0, -1).tolist()
        
        start_time = time.time_ns()  # Get the start time in nanoseconds
        bubble_sort(arr)             # Sort the array
        end_time = time.time_ns()    # Get the end time in nanoseconds
        
        elapsed_time = end_time - start_time
        times.append(elapsed_time)
        print(f"Time taken to sort an array of size {size}: {elapsed_time} nanoseconds")
    return times

# --- 2. Running Bubble Sort Time Measurement and Plotting ---

# Define a range of input sizes to test Bubble Sort
# Note: Bubble Sort becomes very slow for large inputs (e.g., > 10,000)
bubble_sort_input_sizes = [10, 100, 500, 1000, 2000, 3000, 5000, 10000] # Adjusted maximum for faster execution

# Measure the time taken for each input size
bubble_sort_times = measure_time_bubble_sort(bubble_sort_input_sizes)

# Plotting the results of Bubble Sort's time complexity
plt.figure(figsize=(10, 6))
plt.plot(bubble_sort_input_sizes, bubble_sort_times, marker='o', linestyle='-', color='b')
plt.xlabel('Input Size (n)')
plt.ylabel('Time Taken (nanoseconds)')
plt.title('Experimental Time Complexity of Bubble Sort (Worst Case)')
plt.grid(True)
plt.show()

```

Understanding Time Complexity: A Deep Dive into Theoretical Intuition
When we talk about time complexity, we are not measuring time in seconds, milliseconds, or nanoseconds. This is a crucial starting point. Instead, we are measuring the amount of work an algorithm does, expressed as a function of the input size.

1. Counting Operations, Not Clock Time
The Intuition: Imagine an algorithm as a sequence of very basic, fundamental steps. These steps include:

Arithmetic operations (addition, subtraction, multiplication, division)

Comparisons (e.g., if a > b)

Assignments (e.g., x = 5)

Accessing an element in an array (e.g., arr[i])

Each of these is considered a "unit operation" or "elementary operation." We assume, for theoretical analysis, that each unit operation takes a constant amount of time, regardless of the machine or specific values involved.

Why we do this:

Machine Independence: The actual clock time an operation takes varies wildly between different CPUs, memory speeds, programming languages, compilers, operating system loads, and even environmental factors like temperature. If we measured clock time, an algorithm's "time complexity" would be different every time you ran it on a different setup, making comparisons impossible.

Predictability: By counting abstract operations, we get a measure that is inherent to the algorithm itself, not to the environment it's running in. This allows us to predict how its performance will scale, even if we don't know the exact real-world execution time.

2. The Power of Large Inputs (Asymptotic Analysis)
The Intuition: When evaluating an algorithm, we are most interested in how it performs when pushed to its limits. How does it handle a very large amount of data?

Real-World Relevance: In practical applications, algorithms often process millions or billions of data points (e.g., Google searching the web, Netflix recommending movies, banks processing transactions). Performance on small inputs is rarely the bottleneck.

Identifying the "Growth Trend": For small inputs, minor differences or even "bad" algorithms might seem fast. But as the input size (n) grows, the true nature of the algorithm's efficiency, its "growth trend," becomes apparent.

4. Focusing on the Worst Case
The Intuition: When we analyze an algorithm, we want to provide a reliable guarantee about its performance. We want to know the absolute maximum amount of work it might have to do.

Robustness and Guarantees: In engineering, especially for critical systems, you need to know the limits. If your sorting algorithm takes X time in the best case but 100X in the worst case, and the worst case can realistically occur, then your system must be designed to handle 100X time.

Risk Management: Assuming best-case performance is risky. If a system's functionality depends on a search completing within a certain time, and you've only analyzed its best-case (e.g., element found immediately), you're setting yourself up for failure if the element is typically found at the very end of a list.

Predictable Behavior: The worst-case complexity provides an upper bound on the time an algorithm will take. This allows developers to make informed decisions about resource allocation and system design, ensuring that the software remains responsive even under adverse conditions.

```python
# --- 3. Visualizing the Impact of Dominant Terms (n^2 vs n^2 + n) ---

# Define a range of 'n' values for plotting
# Start with a smaller range to see the initial difference, then expand to show diminishing gap
n_values_small = np.arange(1, 40)
n_values_medium = np.arange(1, 400)
n_values_large = np.arange(1, 4000) # Further increased to clearly show convergence

def plot_dominant_term_impact(n_values, title):
    """
    Helper function to plot n^2 and n^2 + n to show how lower-order terms diminish.
    """
    n_squared = n_values ** 2
    n_plus_n_squared = n_values ** 2 + n_values

    plt.figure(figsize=(10, 6))
    plt.plot(n_values, n_squared, label='$n^2$', color='red')
    plt.plot(n_values, n_plus_n_squared, label='$n^2 + n$', linestyle='--', color='blue', alpha=0.7)
    plt.xlabel('Input Size (n)')
    plt.ylabel('Value')
    plt.title(f'Impact of Lower-Order Terms: {title}')
    plt.legend()
    plt.grid(True)
    plt.show()

print("\n--- Visualizing Impact of Lower-Order Terms ($n^2$ vs $n^2 + n$) ---")
```

```python
# Plot for small 'n' to show initial difference
plot_dominant_term_impact(n_values_small, 'Small Input Range (1-40)')

# Plot for medium 'n' to show curves getting closer
plot_dominant_term_impact(n_values_medium, 'Medium Input Range (1-400)')

# Plot for large 'n' to show curves becoming indistinguishable
plot_dominant_term_impact(n_values_large, 'Large Input Range (1-4000) - Indistinguishable Curves')
```
